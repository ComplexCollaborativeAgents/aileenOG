```python
def convert_raw_data_to_dataframe_new(filelist):
    dataframe = pandas.DataFrame(columns = ['run_number', 'task_number', 'memory_steps', 'gscore', 'sscore'])
    run_number = 0
    for file in filelist:
        with open(file) as csvfile:
            reader = csv.DictReader(csvfile)
            for row in reader:
                task_number = row[None][0]
                memory_steps = row[None].count('cm')
                sscore = row[None][-2]
                gscore = row[None][-3]
                new_row = {'run_number': int(run_number), 'task_number': int(task_number), 'memory_steps': int(memory_steps), 'gscore':int(gscore), 'sscore': int(sscore)}
                dataframe = dataframe.append(new_row, ignore_index = True)
        run_number = run_number + 1
    dataframe['run_number'] = pandas.to_numeric(dataframe['run_number'])
    dataframe['task_number'] = pandas.to_numeric(dataframe['task_number'])
    dataframe['memory_steps'] = pandas.to_numeric(dataframe['memory_steps'])
    dataframe['gscore'] = pandas.to_numeric(dataframe['gscore'])
    dataframe['sscore'] = pandas.to_numeric(dataframe['sscore'])
    
    return dataframe
```

```python
def plot_interactive_learning_results(dataset, sdf):
    sns.set(rc={'figure.figsize':(11.7,8.27)}, font="Inconsolata",font_scale=2, style="ticks")
    fig, ax = plt.subplots()
    sns.barplot(x=dataset.task_number, y=dataset.memory_steps, color="mediumaquamarine", ax=ax, alpha=0.5, errcolor='mediumaquamarine')
    ax2 = ax.twinx()
    sns.lineplot(x=dataset.task_number, y=dataset.score, color="coral",  ax=ax2)
    mean = "specifcity exam mean = " + str(sdf['s_score'].mean())
    count = "averaged over " + str(sdf['run_number'].count()) + " runs"
    plt.text(3.5, 5.8,  count, horizontalalignment='left', size='small', color='black', weight='semibold')
    plt.text(3.5, 5.5, mean, horizontalalignment='left', size='small', color='black', weight='semibold')
```

```python
def convert_raw_data_to_dataframe(filelist):
    dataframe = pandas.DataFrame(columns = ['run_number', 'task_number', 'memory_steps', 'score'])
    run_number = 0
    sdf = pandas.DataFrame(columns=['run_number', 's_score'])
    for file in filelist:
        with open(file) as csvfile:
            reader = csv.DictReader(csvfile)
            for row in reader:
                if 'specificity' not in row[None][0]:
                    task_number = row[None][0]
                    memory_steps = row[None].count('cm')
                    score = row[None][-1]
                    new_row = {'run_number': int(run_number), 'task_number': int(task_number), 'memory_steps': int(memory_steps), 'score':int(score)}
                    dataframe = dataframe.append(new_row, ignore_index = True)
                else:
                    new_row = {'run_number': run_number, 's_score': re.findall("\d+", row[None][0])[0]}
                    sdf = sdf.append(new_row, ignore_index = True)
        run_number = run_number + 1
    dataframe['run_number'] = pandas.to_numeric(dataframe['run_number'])
    dataframe['task_number'] = pandas.to_numeric(dataframe['task_number'])
    dataframe['memory_steps'] = pandas.to_numeric(dataframe['memory_steps'])
    dataframe['score'] = pandas.to_numeric(dataframe['score'])
    sdf['run_number'] = pandas.to_numeric(sdf['run_number'])
    sdf['s_score'] = pandas.to_numeric(sdf['s_score'])
    return dataframe, sdf
```

```python
filelist = (glob.glob("data/visual-word/*.csv"))
visual_dataset, sdf = convert_raw_data_to_dataframe(filelist)
visual_dataset.dtypes, sdf.dtypes
```

```python
plot_interactive_learning_results(visual_dataset, sdf)
```

### New runs

```python
def convert_raw_data_to_dataframe_new(filelist):
    dataframe = pandas.DataFrame(columns = ['run_number', 'task_number', 'memory_steps', 'gscore', 'sscore'])
    run_number = 0
    for file in filelist:
        with open(file) as csvfile:
            reader = csv.DictReader(csvfile)
            for row in reader:
                task_number = row[None][0]
                memory_steps = row[None].count('cm')
                sscore = list(str(row[None][-1]))[1]
                gscore = list(str(row[None][-1]))[0]
                new_row = {'run_number': int(run_number), 'task_number': int(task_number), 'memory_steps': int(memory_steps), 'gscore':int(gscore), 'sscore': int(sscore)}
                dataframe = dataframe.append(new_row, ignore_index = True)
        run_number = run_number + 1
    dataframe['run_number'] = pandas.to_numeric(dataframe['run_number'])
    dataframe['task_number'] = pandas.to_numeric(dataframe['task_number'])
    dataframe['memory_steps'] = pandas.to_numeric(dataframe['memory_steps'])
    dataframe['gscore'] = pandas.to_numeric(dataframe['gscore'])
    dataframe['sscore'] = pandas.to_numeric(dataframe['sscore'])
    
    return dataframe
```

```python
def plot_interactive_learning_results_new(dataset):
    sns.set(rc={'figure.figsize':(11.7,8.27)}, font="Inconsolata",font_scale=2, style="ticks")
    fig, ax = plt.subplots()
    sns.barplot(x=dataset.task_number, y=dataset.memory_steps, color="mediumaquamarine", ax=ax, alpha=0.5, errcolor='mediumaquamarine')
    ax2 = ax.twinx()
    sns.lineplot(x=dataset.task_number, y=dataset.gscore, color="coral",  ax=ax2)
    sns.lineplot(x=dataset.task_number, y=dataset.sscore, color="blueviolet",  ax=ax2)
    count = "averaged over " + str(dataset['run_number'].nunique()) + " runs"
    plt.text(3.5, 5.8,  count, horizontalalignment='left', size='small', color='black', weight='semibold')
```

```python
filelist = (glob.glob("data/new-visual-word/*.csv"))
visual_dataset = convert_raw_data_to_dataframe_new(filelist)
visual_dataset
```
